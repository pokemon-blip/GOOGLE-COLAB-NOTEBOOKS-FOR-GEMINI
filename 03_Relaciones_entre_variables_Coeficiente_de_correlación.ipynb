**Coeficiente de correlación**

El coeficiente de correlación es una medida que calcula la cercanía de asociación de los valores de dos variables en un diagrama de dispersión, empleando para ello una regresión lineal. Los ejes x, y deben ser atributos cuantitativos.
El valor del resultado puede ser cualquier número entre -1 y 1 y dependiendo de este valor, podemos comprobar si dos variables cuantitativas tienen relación o no.
- cor: esta función calcula la correlación entre dos variables.

```
cor(data$salary17_18, data$PTS)
```

También podemos utilizar
- cor.test: esta función calcula la correlación entre dos variables

```
correlation <- cor.test(data$salary17_18, data$PTS)
correlation$estimate
correlation$p.value
```

Para analizar los resultados tenemos que ver el resultado de la prueba (correlación) y también el resultado del valor p. Recordemos que si el valor p es menor a un valor umbral (generalmente 0.05) podemos afirmar que el test realizado es estadísticamente relevante. Por otro lado, el valor de correlación, nos da información sobre la relación entre las variables:

-1.0: Correlación negativa perfecta
-0,8 a -1,0: correlación negativa muy fuerte
-0,6 a -0,8: fuerte correlación negativa
-0,4 a -0,6: correlación negativa moderada
-0,2 a -0,4: correlación negativa débil
0.0 a -0.2: correlación negativa muy débil o no correlación
0.0 a 0.2: correlación positiva muy débil o no correlación
0,2 a 0,4: correlación positiva débil
0,4 a 0,6: correlación positiva moderada
0,6 a 0,8: fuerte correlación positiva
0.8 a 1.0: correlación positiva muy fuerte
1.0: Correlación positiva perfecta

Otra forma de verificar la correlación es dibujar el diagrama de dispersión entre las dos variables y agregar la línea de regresión para ver las distancias de sus puntos.

```
with(data,plot(salary17_18, PTS))
abline(lm(PTS~salary17_18,data=data), col="red" )
```
