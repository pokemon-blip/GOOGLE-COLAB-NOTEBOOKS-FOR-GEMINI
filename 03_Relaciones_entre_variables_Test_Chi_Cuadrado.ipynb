 **Test Chi Cuadrado**

El test de Chi Cuadrado es una herramienta útil para determinar si dos variables categóricas son independientes o no. En este sentido, tenemos que definir las hipótesis nula y alternativa para realizar el test:

- Hipótesis nula (H0): las variables son independientes, es decir, no existe relación entre las dos variables categóricas. Además, no podemos predecir el valor de una de las dos variables conociendo el valor de la otra.

- Hipótesis alternativa (H1): las variables son dependientes, es decir, existe una relación entre las dos variables categóricas. Además, podemos predecir el valor de una de las dos variables conociendo el valor de la otra.

Después de esta breve explicación, vamos a comprobar si dos atributos categóricos son dependientes o independientes. Para el primer test, elegiremos "City" y "Medal". En la siguiente tabla de frecuencias podemos ver la distribución de ambas variables.

```
table(data$City,data$Medal)
```

A continuación, vamos a ejecutar el test Chi Cuadrado
- chisq.test: ejecuta el test Chi Cuadrado entre las columnas seleccionadas por parámetro

```
chiSquaredTest <- chisq.test(data$City,data$Medal)
chiSquaredTest
```

Para comenzar a analizar los resultados de la prueba, comprobaremos el valor p. Estableciendo un umbral para este valor (p <= 0.05), podemos determinar si los resultados de la prueba son estadísticamente significativos. Concretamente, si el valor resultante es superior al umbral, no podemos rechazar la hipótesis nula, es decir, no podemos afirmar que las variables categóricas no sean independientes.

Ahora, vamos a realizar la prueba de chi al cuadrado con las columnas *Season* y *Sport*.

```
table(data$Season,data$Sport)
```

Como podemos ver en la tabla de frecuencias anterior, parece que existe alguna relación entre la temporada y el deporte (por ejemplo, el esquí alpino es un deporte de invierno típico). Para confirmarlo vamos a realizar de nuevo el Test Chi Cuadrado

```
chiSquaredTest <- chisq.test(data$Season,data$Sport)
chiSquaredTest
```

En este caso, si el valor p es menor que el umbral, podemos rechazar la hipótesis nula. Además, si el valor de X cuadrado es alto, significa que las variables son dependientes.
