**Test de ANOVA**

El test de ANOVA estima cómo cambia la variable cuantitativa según los niveles de una o más variables categóricas.
En este cuaderno, primero vamos a comenzar realizando un test de ANOVA de una vía, eso significa que vamos a suponer que tenemos una variable independiente.

- aov: esta función ejecuta el test de ANOVA. El primer parámetro es la fórmula en la que indicamos cuál es la variable independiente y cuál es la dependiente.


```
anovaOneWay <- aov(salary17_18 ~ Pos, data = data)
```

Al ejecutar el comando *summary*, podemos ver diferentes valores extraídos de la prueba:
- Df: grados de libertad de la variable independiente
- Sum Sq: variación entre las medias del grupo y la media total
- Mean Sq: división de "Sum Sq" entre "Df"
- Value F: cuanto mayor es el valor, más probable es la variación causada por la variable independiente
- Pr (> F): muestra la probabilidad de que se haya producido el valor F si la hipótesis nula fuera cierta.

En este punto es necesario explicar qué significa hipótesis nula en este contexto. La hipótesis nula en el Test de ANOVA se corresponde a la situación en la que no hay diferencia entre medias de los grupos en cada nivel de la variable categórica, es decir, las dos variables son independientes.
Por otro lado, la hipótesis alternativa es que las medias de los grupos son diferentes.

```
summary(anovaOneWay)
```

Para empezar a analizar los resultados de la prueba, vamos a fijarnos en la columna Pr (> F). Podemos afirmar que cuando este valor es bajo, el resultado es estadísticamente significativo (aquí es necesario establecer un umbral como p <0.05). En cuanto al valor F, cuanto mayor sea, más diferencia entre las medias (más dependencia).


*Test de Tukey*

El test de ANOVA puede mostrarnos que existen diferencias entre las medias de los grupos, pero no muestra cuáles son las diferencias. Para ello, podemos utilizar el test deTukey.
- TukeyHSD: calcula el test de Tukey. El primer parámetro que recibe es un objeto con el resultado de prueba ANOVA.


```
tukeyOneWay <- TukeyHSD(anovaOneWay)
tukeyOneWay
```

Aquí nuevamente, tenemos que analizar la columna * p adj *. Recuerde el triple anterior (p <0.05) que significa que si el valor es mayor que el umbral, el resultado no es estadísticamente significativo.
Otra forma de mostrar los resultados es usar la función * plot *.

```
plot(tukeyOneWay)
```
