**Tipos de atributos**

Desde un punto de vista estadístico tenemos:

- Cualitativos
    - Atributos nominales
    - Atributos ordinales

- Cuantitativos
    - Atributos discretos
    - Atributos continuos

```
nominalAttribute = data$Sex
ordinalAttribute = data$Medal
discreteAttribute = data$Age
continousAttribute = data$Height
```

```
cat ("Nominal attribute (Sex):\n ")
str(nominalAttribute)
cat ("Ordinal attribute (Medal):\n ")
str(ordinalAttribute)
cat ("Discrete attribute (Age):\n ")
str(discreteAttribute)
cat ("Continuous attribute (Height):\n ")
str(continousAttribute)
```

**Escalas**

- Con atributos nominales:
  - Conteo

```
unique(nominalAttribute)
```

- Con atributos ordinales:
  - Conteo
  - Orden

```
unique(ordinalAttribute)
ordinalAttribute <- factor(ordinalAttribute, levels = c("Bronze","Silver","Gold"))
```

Con atributos discretos:
  - Conteo
  - Orden
  - Resta

```
# Difference between youngest and oldest athlete (ommiting na values)
ageDifference = max(discreteAttribute, na.rm=TRUE)-min(discreteAttribute, na.rm=TRUE)
ageDifference
```

Con atributos continuos:
  - Conteo
  - Orden
  - Resta
  - División

```
# Height average using a division(ommiting na values)
heighAverage =  sum(continousAttribute, na.rm=TRUE)/length(continousAttribute, na.rm=TRUE)
heighAverage
```

**Tablas**
- table: crea una tabla de frecuencias a partir de los valores de una columna

```
table(nominalAttribute)
```

- prop.table: crea una tabla con la distribución de cada valor de una columna

```
prop.table(table(nominalAttribute))
```

- CrossTable: creates una tabla cruzada con los datos de varias columnas (gmodels library).

```
CrossTable(nominalAttribute,ordinalAttribute)
```
