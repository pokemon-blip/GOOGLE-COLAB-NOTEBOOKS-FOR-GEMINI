**Medidas de tendencia central**
- mean: calcula la media aritmética de una columna

```
mean(data$Height, na.rm=TRUE)
```

- median: calcula la mediana de una columna

```
median(data$Age, na.rm=TRUE)
```

- mfv: calcula la moda de una columna

```
mfv(data$Height, na_rm=TRUE)
mfv(data$Sport, na_rm=TRUE)
```

**Medidas de dispersión**

- range: devuelve el valor mínimo y el máximo de una columna



```
range(data$Height, na.rm= TRUE)
```

 - var: calcula la varianza de una columna

```
var(data$Height, na.rm= TRUE)
```

- sd: calcula la desviación típica de una columna

```
sd(data$Height, na.rm= TRUE)
```

- cv: calcula el coeficiente de variación de una columna

```
cv(data$Height, na.rm= TRUE)
```

- En el caso del coeficiente de rango, podemos aplicar la formula directamente (max-min)/(max+min)


```
coefficientOfRange = (max(data$Height, na.rm=TRUE)-min(data$Height, na.rm=TRUE))/(max(data$Height, na.rm=TRUE)+min(data$Height, na.rm=TRUE))
coefficientOfRange
```

**Otras medidas:**
  - skewness: calcula la asimetría de una columna

```
skewness(data$Height, na.rm= TRUE)
skewness(data$Age, na.rm= TRUE)
```

- kurtosis: calcula la curtosis de una columna

```
kurtosis(data$Height, na.rm= TRUE)
kurtosis(data$Age, na.rm= TRUE)
```

Podemos ver también estas medidas con el comando *hist*

```
hist(data$Height)
hist(data$Age)
```

**Extra**

- summary: este comando devuelve varias medidas dependiendo del tipo de variable que le pasemos.
Se puede utilizar con una única columna o con todo el dataframe.

```
summary(data)
summary(data$Height)
```
